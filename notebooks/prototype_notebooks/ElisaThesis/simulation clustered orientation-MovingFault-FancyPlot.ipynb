### Moving fault data:

```python
# Initial data
select_i = geo_data.interfaces['formation'] == 'fault_NWSE'
x_i = geo_data.interfaces[select_i]['X']
y_i = geo_data.interfaces[select_i]['Y']

select_o = geo_data.orientations['formation'] == 'fault_NWSE'
x_o = geo_data.orientations[select_o]['X']
y_o = geo_data.orientations[select_o]['Y']
angle = 45
```

```python
for i in range(5):
    distance = np.random.normal(0, 1000)
    # Changing the dataframes:

    geo_data.interfaces['X'][select_i] = x_i.apply(lambda x: x + distance * np.cos(angle * np.pi / 180))
    geo_data.interfaces['Y'][select_i]  = y_i.apply(lambda x: x + distance * np.sin(angle * np.pi / 180))

    geo_data.orientations['X'][select_o] = x_o.apply(lambda x: x + distance * np.cos(angle * np.pi / 180))
    geo_data.orientations['Y'][select_o] = y_o.apply(lambda x: x + distance * np.sin(angle * np.pi / 180))
    # You need to use interp_data.update_interpolator(geo_data) before computing!!!!!!!!!!!!!!!!!!
```

```python
def calculate_ie_masked(self, prob):
    ie = np.zeros_like(prob[0])
    for l in prob:
        pm = np.ma.masked_equal(l, 0)  # mask where prob is 0
        ie -= (pm * np.ma.log2(pm)).filled(0)
return ie
```
